```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
flips = np.random.binomial(1,0.5,1000)
```

```python
heads = flips.sum()
```

```python
heads
```

```python
fig, ax = plt.subplots()

ax.bar([0,1],[1000 - heads,heads])

ax.set_xticks([0,1])
ax.set_title("flipping a coin 1000 times")
ax.set_ylabel("no heads")
ax.set_label("Count")
```
